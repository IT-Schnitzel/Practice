1.3.4 Задача: Создать матрицу с 0 внутри, и 1 на границах

```python
import numpy as np
print("Введите размер матрицы m*n через пробел")
m,n = list(map(int,input().split()))
for i in range (m):
  if (i == 0 or i == m-1):
    print(np.ones(n,dtype = int))
  else:
    A = np.zeros(n,dtype = int)
    A[0] = 1
    A[n-1] = 1
    print(A)
```

2.3.1 Задача: Найдите евклидово расстояние между двумя Series (точками) a и b, не
используя встроенную формулу.


```python
import random
import math
import numpy as np
import pandas as pd
s = random.randint(1,23)
A = [0]*s
B = [0]*len(A)
sum = 0
for i in range (len(A)):
  A[i] = random.random()
  B[i] = random.random()
  sum += abs(A[i] - B[i])**2
#print(A)
#print(B)
a = pd.Series(A)
b = pd.Series(B)
print(math.sqrt(sum))
print(np.linalg.norm(a-b))
```

3.3.2 Задание Задача:
Загрузить фрейм данных по ссылке:
https://raw.githubusercontent.com/akmand/datasets/master/iris.csv.
Необходимо выполнить нормализацию первого числового признака
(sepal_length_cm) с использованием минимаксного преобразования, а
второго (sepal_width_cm) с задействованием z-масштабирования.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd

url = "https://raw.githubusercontent.com/akmand/datasets/master/iris.csv"
dataframe = pd.read_csv(url)

min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

dataframe['sepal_length_cm'] = min_max_scaler.fit_transform(dataframe[['sepal_length_cm']])
dataframe['sepal_width_cm'] = standard_scaler.fit_transform(dataframe[['sepal_width_cm']])

dataframe.head(20)
```
